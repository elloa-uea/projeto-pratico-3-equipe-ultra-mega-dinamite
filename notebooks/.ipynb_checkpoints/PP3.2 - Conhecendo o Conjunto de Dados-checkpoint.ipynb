### Visualizando a base de dados

1. Baseando-se nos fundamentos de visualização de dados abordados na disciplina, plote cada um dos atributos preditores de maneira a enfatizar a sua distribuição, tendência central e dispersão

```python
def generate_statistics(column):
    return df[column].agg(np.median),df[column].agg(np.mean),df[column].mode()[0],df[column].agg(np.std),df[column].agg(np.var)
```

```python
def create_scatter(column,title,xlabel,s):
    # GRÁFICO DE DISPERSÃO
    plt.subplot(2, 2, 2)
    data = df[column].value_counts()
    plt.title(title+'/Ocorrências', fontsize=15)
    plt.ylabel('Número de ocorrências', fontsize=12)
    plt.xlabel(xlabel, fontsize=12)
    plt.scatter(data.index,data.values,s=s)
    plt.show()
```

```python
def create_graph(title='',ylabel='Número de instâncias',xlabel='Distância em metros',column='',density=150,color='tab:blue',s=2,ytext=3000):
    # GRÁFICO DE DISTRIBUIÇÃO + DISPERSÃO
    plt.figure(figsize=(16, 9))
    plt.subplot(2, 2, 1)
    median,mean,mode,std,var = generate_statistics(column)
    plt.title(title, fontsize=15)
    plt.xlabel(xlabel,fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    me = plt.axvline(x=median, ymin=0, color='orange')
    ma = plt.axvline(x=mean, ymin=0, color='cyan', linestyle='dashed')
    extra = mpatches.Rectangle((0, 0), 1, 1, fc="w", fill=False, edgecolor='none', linewidth=0)
    plt.annotate('Mode', xy=(mode, 0), xytext=(mode, ytext),arrowprops=dict(arrowstyle="-|>"))
    plt.legend([me,ma,extra],('Median: {:.2f}'.format(median),
                              'Mean: {:.2f}'.format(mean),
                             "Mode: {}\nStd: {:.2f}\nVar: {:.2f}".format(mode,std,var)))
    plt.hist(df[column], density, facecolor=color)
    plt.grid(axis='y')
    create_scatter(column,title,xlabel,s)
```

```python
# Elevação em metros
create_graph(title='Elevação em metros',column='Elevation')
```

```python
# Aspecto Azimuth
create_graph(title='Aspecto em graus azimute',xlabel='Azimuth\u00b0',column='Aspect',ytext=2000)
```

```python
# Inclinação de iluminação
create_graph(title='Inclinação de iluminação em graus',xlabel='Inclinação\u00b0',column='Slope',density=67,ytext=5000)
```

```python
# Distância horizontal para fontes de água
create_graph(title='Distância horizontal para fontes de água',column='Horizontal_Distance_To_Hydrology',density=80,ytext=5000)
```

```python
# Distância vertical para fontes de água
create_graph(title='Distância vertical para fontes de água',column='Vertical_Distance_To_Hydrology',density=80,ytext=10000)
```

```python
# Distância horizontal para rua
create_graph(title='Distância horizontal para rua',column='Horizontal_Distance_To_Roadways',density=80)
```

```python
# Distância para origens de incêndio
create_graph(title='Distância para origens de incêndio',column='Horizontal_Distance_To_Fire_Points')
```

```python
# Iluminação às 9h
create_graph(title='Índice Hillshade às 9h',xlabel='Iluminação às 9h',column='Hillshade_9am',density=255)
# Iluminação ao meio-dia
create_graph(title='Índice Hillshade às 12h',xlabel='Iluminação às 12h',column='Hillshade_Noon',density=255)
# Iluminação às 15h
create_graph(title='Índice Hillshade às 15h',xlabel='Iluminação às 15h',column='Hillshade_3pm',density=255,ytext=2000)
```

```python
def create_graph_soil_or_wilderness(column, n, title, xlabel, ylabel='Número de instâncias'):
    array_occurrences = []
    for i in range(1,n+1):
        occurrences = len(df.loc[df[column+str(i)] == 1])
        array_occurrences.append(occurrences)
    df_attribute = pd.DataFrame(list(zip(np.arange(1, n+1), array_occurrences)),
                                columns=[column, 'y'])
    # GRÁFICO DE DISTRIBUIÇÃO + DISPERSÃO
    plt.figure(figsize=(16, 9))
    plt.subplot(2, 2, 1)
    plt.title(title+' / Instâncias', fontsize=15)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel,fontsize=12)
    plt.bar(df_attribute[column], df_attribute['y'])
    plt.subplot(2, 2, 2)
    plt.title(title+'/Instâncias', fontsize=15)
    plt.ylabel(ylabel, fontsize=12)
    plt.xlabel(xlabel, fontsize=12)
    plt.scatter(df_attribute[column],df_attribute['y'],s=30)
    plt.show()
```

```python
# Tipo de Solos
create_graph_soil_or_wilderness('Soil_Type',40,'Designação do tipo de solo','Tipo de solo')
```

```python
# Área Selvagem
create_graph_soil_or_wilderness('Wilderness_Area',4,'Designação de área selvagem','Tipo de área selvagem')
```
